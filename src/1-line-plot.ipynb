### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt  
import numpy as np  

# Generate x values  
x = np.linspace(0, 10, 100)  

# Calculate y values (square of x)  
y = x**2  

# Create the plot  
plt.plot(x, y)  

# Add title and labels  
plt.title("Line Plot of y = x^2")  
plt.xlabel("x")  
plt.ylabel("y")  

# Add grid  
plt.grid(True)  

# Show the plot  
plt.show()  
```
